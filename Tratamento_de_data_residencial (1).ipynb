```python
imoveis_comercial = ['Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão', 'Galpão/Depósito/Armazém', 'Casa Comercial', 'Terreno Padrão', 'Box/Garagem', 'Loja Shopping/ Ct Comercial', 'Studio', 'Hotel', 'Indústria']
```

```python
df_residencial = df.query('Tipo not in @imoveis_comercial')
df_residencial
```

```python
df_residencial.isnull().sum()
```

```python
df_residencial.fillna(0, inplace=True)
df_residencial
```

```python
df_residencial.query('Valor == 0 | Condominio == 0')
```

```python
df_residencial_washed = df_residencial.query('Valor != 0 and Condominio != 0 and Area != 0 and Quartos != 0 and Tipo != "Loteamento/Condomínio"')
df_residencial_washed
```

```python
df_residencial_only_valor = df_residencial_washed.groupby('Tipo')[['Valor']].mean(numeric_only=True)
df_residencial_only_valor.plot(kind='bar',figsize=(14, 10), title='Valor Médio dos Imóveis por Tipo', color = 'Purple')
```

```python
df_residencial_washed.to_csv('data_frame_residencial.csv', sep=';', index=False)
```
